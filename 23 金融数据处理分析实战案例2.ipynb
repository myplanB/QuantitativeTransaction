```python
import tushare as ts  # 引入tushare包
hs300 = ts.get_hs300s() # 获取沪深300的股票数据
hs300
```

```python
hs300 = hs300['code'].tolist()  #  转换成列表
hs300[:10]
```

```python
# 获取基本面数据
stock_basics = ts.get_stock_basics()
type(stock_basics)
stock_basics
```

```python
# 从基本面数据种获取相关信息,获取hs300的
stock_basics.reset_index(inplace=True) # 重置DateFrame的索引
data1 = stock_basics.loc[stock_basics['code'].isin(hs300),  # 行标签
                        ['code','name','industry','pe','pb','esp','rev','profit']]  #列标签

data1.columns = ['代码','名称','行业','PE','PB','EPS','收入%','利润%']
data1.head()
```

```python
#获取盈利能力数据
stock_profit = ts.get_profit_data(2017,1)
```

```python
data2 = stock_profit.loc[stock_profit['code'].isin(hs300),  #行标签
                         ['code','roe','gross_profit_rate','net_profit_ratio']] # 列标签
data2.columns = ['代码','ROE','毛利率','净利率']
data2 = round(data2,2)  # 保留2位小数
data2.head()
```

```python
# 获取成长能力数据
stock_growth = ts.get_growth_data(2019,1)
stock_growth
```

```python
data3 = stock_growth.loc[stock_growth['code'].isin(hs300),  # isin： 是否存在
                        ['code','nprg']]
data3.columns = ['代码','N增长率']
data3 = round(data3,2)  # 实现4舍5入
data3.head()
```

```python
from functools import reduce
import pandas as pd

merge = lambda x,y: pd.merge(x,y,how='left',on='代码')
data = reduce(merge,[data1,data2,data3])
data.drop_duplicates(inplace=True)  # 去掉重复数据
data.head()
```

```python
data['估值系数'] = data['PE'] * data['PB']
data = round(data,2)
data.head()
```

```python
data_filter = data.loc[(data['估值系数']<60) & (data['ROE']>5),
                      ['代码','名称','PE','PB','估值系数','ROE','收入%']]
data_filter.head()
```

```python
print('筛选结果共 %d 只个股' % len(data_filter))
```

```python
# 按照选定字段对数据进行排序
data_filter.sort_values(['估值系数'],ascending=True,inplace=True)  # sort_values 是数值进行排序
data_filter.head(10)
```

```python
# 数据分类
def map_func(x):
    if x['ROE'] > 5:
        return '高成长'
    elif x['ROE'] >= 0 and x['ROE'] <= 5:
        return '低成长'
    elif x['ROE'] < 0:
        return '亏损'
```

```python
data['成长性'] = data.apply(map_func,axis=1)
data.tail(10)
```

```python
# 对高成长分类，按照“估值系数”做升序排列
data_growth = data[data['成长性'] == '高成长'].sort_values(['估值系数'],ascending=True)
data_growth.head(10)
```

```python
def group_func(df):
    return df.sort_values(['估值系数'],ascending=True)[:2]

data_grouped = data.groupby('成长性').apply(group_func)
data_grouped
```
