```python
# 合并操作
# group、concate、join、merge方法
# mutil-index 层次化索引
!pip3 install seaborn

import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# 生成DataFrame
period = pd.date_range('2017-1-1',periods=10000,freq='D')

df = pd.DataFrame(
    np.random.randn(10000,4),
    columns=['Data1','Data2','Data3','Data4'],
    index=period
)
```

```python
df.head()
```

```python
# 在dataframe上增加一列
df['group1'] = np.random.choice(['A','B','C','D'],10000)
df.tail()
```

```python
# 将数据按照‘group1’列进行分组
grouped = df.groupby('group1')
```

```python
type(grouped)
```

```python
grouped.size()
```

```python
# 打印出各类信息
grouped.max()
```

```python
grouped.mean()
```

```python
grouped.describe()
```

```python
# 对数据进行转置
np.transpose(df)
```

```python
# 对已经分组的数据，取其中一组数据，例如取A组数据
grouped.get_group('A').head()
```

```python
grouped.get_group('B').tail()
```

```python
df['group2'] = np.random.choice(['Python','C++','Java'],10000)
```

```python
df.head()
```

```python
# 连续对两列数据进行分组
grouped = df.groupby(['group1','group2'])
```

```python
grouped.size()
```

```python
# 自定义聚合方式
grouped.agg([np.mean,np.sum,np.max,np.min])
```
